```python
import numpy as np
from numpy import arange
from pylab import plot,xlabel,ylabel,show
import matplotlib.pyplot as plt
```

```python
# Write Code Here
# Write Code Here


# Constants

#Set initial number of atoms:


NRn = 2500
NPo = 0
NPb214 = 0
NBi = 0
NTl=0
NPb207=0
total_atoms=NRn+NPo+NPb214+NBi+NTl+NPb207
#Set Tau values
tau_Rn = 2.22*24*60*60
tau_Po = 3.1*60
tau_Pb214 = 26.8*60
tau_Bi = 19.9*60
tau_Tl=1.3*60

h = 1.0                    # Size of time-step in seconds

#calculate probability of decay
p_Rn = 1 - 2**(-h/tau_Rn)
p_Po = 1 - 2**(-h/tau_Po)
p_Pb214 = 1 - 2**(-h/tau_Pb214)
p_Bi = 1 - 2**(-h/tau_Bi)
p_Tl=1 - 2**(-h/tau_Tl)

tmax = 2000000
#calculate probability of decay into which new atom
p_Po_to_Pb=0.0002
p_Po_to_Bi = 0.9998

p_Bi_to_Tl=0.997
p_Bi_to_Pb=0.003

# Lists of plot points
# make time array
# make empty array to store # of each atom at each time step
Rn_points = []
Po_points = [] 
Pb214_points = []
Bi_points = []
Tl_points = []
Pb207_points=[]
total_points = []
t=0
count_var=0

A=0
B=0
R=0
Z=0
# Main loop
while NPb207 < total_atoms:
    count_var+=1
    #Append starting number of each atom
    Rn_points.append(NRn )
    Po_points.append(NPo)
    Pb214_points.append(NPb214 )
    Bi_points.append(NBi)
    Tl_points.append(NTl)
    Pb207_points.append(NPb207)
    total_points.append(NRn+NPo+NPb214+NBi+NTl+NPb207)

    
    # Calculate the number of atoms that decay
    decay_from_Rn=0
    decay_to_Po=0
    decay_to_Pb214=0
    decay_to_Bi=0
    decay_to_Tl=0
    decay_to_Pb207=0

    for i in range (NRn):
        if NRn>=1 and np.random.random() < p_Rn:
            decay_to_Po+=1
            decay_from_Rn+=1
            R+=1

    for j in range(NPo):
        # determine number of Po atoms that decay:
        if  NPo>=1 and np.random.random() < p_Po:
            decay_to_Po-=1
            # determine how much decays to Pb
            if np.random.random() < p_Po_to_Pb:
                decay_to_Pb214 += 1
                A+=1
                
            # If it doesn't decay to Pb, but does decay, it must decay to Bi
            else:
                decay_to_Bi +=1
                B+=1
    for k in range (NPb214):
        if NPb214>=1 and np.random.random()<p_Pb214:
            decay_to_Bi+=1
            decay_to_Pb214-=1
            Z+=1
    for l in range (NBi):
        if NBi >= 1 and np.random.random() < p_Bi:
            decay_to_Bi-=1
            if np.random.random() < p_Bi_to_Tl:
                decay_to_Tl+=1    
                A+=1
            else:
                decay_to_Pb207+=1
                R+=1
    for m in range (NTl):
        if NTl >=1 and np.random.random() < p_Tl:
            decay_to_Pb207+=1
            decay_to_Tl-=1
            B+=1
    if count_var % 1000 == 0:
        print(f"{100*NPb207/(total_atoms)}% complete \r",end='\r', flush=True)                      
    NRn -= decay_from_Rn
    NPo += decay_to_Po
    NPb214 += decay_to_Pb214
    NBi += decay_to_Bi
    NTl += decay_to_Tl
    NPb207 += decay_to_Pb207
    total_atoms=NRn+NPo+NPb214+NBi+NTl+NPb207
    t+=h
tpoints = arange(0.0,t,h) 






```

```python

# Make the graph
fig = plt.figure(figsize=(12,7))
ax = fig.add_subplot(111)
ax.plot(tpoints,Rn_points, c='g', label = "Rn") # plot Tl vs. time
ax.plot(tpoints,Po_points, c='r', label = "Po")
ax.plot(tpoints,Pb214_points, c='k', label = "Pb214")
ax.plot(tpoints,Bi_points, c='b', label = "Bi")
ax.plot(tpoints,Tl_points, c='m', label = "Tl")
ax.plot(tpoints,Pb207_points, c='orange', label = "Pb207")
ax.plot(tpoints,total_points, c='cyan', label = "total_atoms")
# add the labels
ax.set_xlabel("Time (s)",size=20)   # allows LaTeX style formating
ax.set_ylabel("Number of atoms",size=20)   # allows LaTeX style formating
# increase the size of the numbers on the axes
ax.xaxis.set_tick_params(labelsize=20)
ax.yaxis.set_tick_params(labelsize=20)
plt.legend()
plt.show()
```

```python
def energy (A,B,R,Z):
    return 4*A, 3*B, 9*R, 8*Z
def thickness(A, B, R, Z):
    eA, eB, eR, eZ = energy(A,B,R,Z)
    return eA/3000, eB/5000, eR/6000, eZ/1000
def cost (A, B, R, Z):
    tA, tB, tR, tZ = thickness(A,B,R,Z)
    return tA*3, tB*5, tR*200, tZ*50


print(energy(A,B,R,Z))
print(thickness(A,B,R,Z))
print(cost(A,B,R,Z))

```
